```python
def extract_price(obj):
    """
    Returns price if obj is dict or number, else None
    """
    if isinstance(obj, dict):
        return obj.get("price")
    if isinstance(obj, (int, float)):
        return obj
    return None
```

```python
def _log_price_diff(logs, category, service_name, contract_data, catalog_data):
    contract_price = extract_price(contract_data)
    catalog_price = extract_price(catalog_data)

    if contract_price is None or catalog_price is None:
        return

    price_diff = contract_price - catalog_price

    if price_diff < 0:
        status = "UNDER_PRICED"   # leakage
    elif price_diff > 0:
        status = "OVER_PRICED"
    else:
        status = "MATCHED"

    logs.append({
        "category": category,
        "service": service_name,
        "unit": (
            contract_data.get("unit")
            if isinstance(contract_data, dict)
            else catalog_data.get("unit")
            if isinstance(catalog_data, dict)
            else None
        ),
        "catalog_price": catalog_price,
        "contract_price": contract_price,
        "price_diff": price_diff,
        "pricing_status": status
    })
```

```python
def compare_contract_vs_catalog(contract_result, catalog):
    logs = []

    CATEGORY_MAP = {
        "Cash Management": "cash_management",
        "Domestic Payments": "domestic_payments",
        "International Payments": "international_payments",
        "Trade Finance": "trade_finance",
        "Digital Services": "digital_services"
    }

    for contract_category, contract_services in contract_result.items():
        if contract_category not in CATEGORY_MAP:
            continue

        catalog_category_key = CATEGORY_MAP[contract_category]
        catalog_services = catalog.get(catalog_category_key, {})

        for service_name, contract_data in contract_services.items():

            # Nested services (standing_instruction, bank_guarantee)
            if isinstance(contract_data, dict) and "price" not in contract_data:
                for sub_service, sub_data in contract_data.items():
                    catalog_sub = catalog_services.get(service_name, {}).get(sub_service)
                    if not catalog_sub:
                        continue

                    _log_price_diff(
                        logs,
                        contract_category,
                        f"{service_name}.{sub_service}",
                        sub_data,
                        catalog_sub
                    )
                continue

            catalog_data = catalog_services.get(service_name)
            if not catalog_data:
                continue

            _log_price_diff(
                logs,
                contract_category,
                service_name,
                contract_data,
                catalog_data
            )

    return logs
```

```python
def _log_price_diff(logs, category, service_name, contract_data, catalog_data):
    contract_price = contract_data.get("price")
    catalog_price = catalog_data.get("price")

    # Handle percentage pricing separately
    if contract_price is None or catalog_price is None:
        return

    price_diff = contract_price - catalog_price

    if price_diff < 0:
        status = "UNDER_PRICED"   # leakage risk
    elif price_diff > 0:
        status = "OVER_PRICED"
    else:
        status = "MATCHED"

    logs.append({
        "category": category,
        "service": service_name,
        "unit": contract_data.get("unit") or catalog_data.get("unit"),
        "catalog_price": catalog_price,
        "contract_price": contract_price,
        "price_diff": price_diff,
        "pricing_status": status
    })
```

```python
def compute_contract_severity(pricing_logs):
    CATEGORY_WEIGHTS = {
        "Cash Management": 1.0,
        "Domestic Payments": 1.0,
        "International Payments": 1.2,
        "Trade Finance": 1.5,
        "Digital Services": 0.8
    }

    total_weighted_severity = 0.0
    total_weight = 0.0

    for row in pricing_logs:
        catalog_price = row["catalog_price"]
        contract_price = row["contract_price"]

        # Skip if price missing or percentage handled elsewhere
        if catalog_price is None or contract_price is None:
            continue

        raw_severity = (contract_price - catalog_price) / catalog_price

        # Cap to [-1, 1]
        severity = max(-1.0, min(1.0, raw_severity))

        weight = CATEGORY_WEIGHTS.get(row["category"], 1.0)

        total_weighted_severity += severity * weight
        total_weight += weight

    if total_weight == 0:
        return 0.0

    return round(total_weighted_severity / total_weight, 3)
```
